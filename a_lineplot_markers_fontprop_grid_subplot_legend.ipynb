<h1>Matplotlib Library:</h1>
<ul>
    <li>Matplotlib is a low level graph plotting library in python that serves as a visualization utility.</li>
    <li>Matplotlib was created by John D. Hunter.</li>
    <li>Matplotlib consists of two submodules. They are:</li>
</ul>
<ol>
    <li>Pyplot</li>
    <li>Pylab</li>
</ol>
<ul>
    <li>Most of the matplotlib utilities lies under the pyplot submodule.</li>
    <li>Types of charts that are available in matplotlib</li>
</ul>
<ol>
    <li>Line Chart</li>
    <li>Scatter Chart</li>
    <li>Bar Chart</li>
    <li>Histogram Chart</li>
    <li>Pie Chart</li>
</ol>

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h3>1. Line chart:</h3>

```python
# Creating a line plot
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

# Drawing points in a diagram
plt.plot(xpoints, ypoints)

# Displaying the plot
plt.show()
```

```python
# If you specify only one parameters it takes on y-axis
xpoints = np.array([1, 2, 6, 8])

plt.plot(xpoints)

plt.show()
```

```python
x = [5, 2, 7]
y = [2, 16, 4]

# Adding font properties to the lines and markers
plt.plot(x, y, marker = "*", ms = 20, mec = "magenta", mfc = "cyan", ls = "-.", lw = 3, color = "g")

plt.show()
```

```python
x1 = np.arange(0, 10, 2)
y1 = x1 ** 2
print(x, y)

plt.plot(x1, y1, marker = "*", ms = 20, mec = "k", mfc = "y")

# Adding title, x-label, y-label to the plot
plt.title("Example for markers")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
points = np.array([3, 8, 1, 10])

# Using short cut string notation parameter to specify the marker
plt.plot(points, "o:r")

plt.show()
```

```python
# Plotting Without a Line
xx = np.array([1, 8])
yy = np.array([3, 10])

plt.plot(xx, yy, 0)

plt.show()
```

<h4>Muliple Lines:</h4>
<ul>
    <li>You can plot as many lines as you like by simply adding more plot() functions or by adding the points for the x-axis and y-axis for each line in the same plot() function.</li>
</ul>

```python
# With multiple Plot Funtions
x1 = np.arange(0, 10, 2)
y1 = x1 ** 2
x2 = np.array([10, 20, 30, 50, 30])
y2 = [i ** 2 for i in x2]

plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y1)

plt.show()
```

```python
# With one plot function
x1 = np.arange(0, 10, 2)
y1 = x1 ** 2
x2 = np.array([10, 20, 30, 50, 30])
y2 = [i ** 2 for i in x2]

plt.plot(x1, y1, x1, y2, x2, y1)

plt.show()
```

<h4>Font properties for Title and Labels:</h4>

```python
x1 = np.arange(0, 10, 2)
y1 = x1 ** 2
print(x, y)

f1 = {
    "family":"serif",
    "color":"magenta",
    "size":30
}
f2={
    "family":"serif",
    "color":"cyan",
    "size":30
}

plt.plot(x1, y1, marker = "*", ms = 20, mec = "k", mfc = "y")

plt.title("Example for markers", fontdict = f1, loc = "right")
plt.xlabel("x-axis", fontdict = f2, loc = "left")
plt.ylabel("y-axis", fontdict = f2, loc = "top")

plt.show()
```

<h4>Grid lines:</h4>
<ul>
    <li>With Pyplot, you can use the grid() function to add grid lines to the plot.</li>
</ul>

```python
x = [2, 3, 5, 4, 6, 7]
y = [i * 2 for i in x]

plt.plot(x, y, marker = "o")
plt.grid(axis = "both", ls = "dashdot", lw = 1, c = "r")

plt.show()
```

<h4>Subplot:</h4>
<ul>
    <li>With the subplot() function you can draw multiple plots in one figure.</li>
</ul>

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 0])

plt.subplot(2, 1, 1)
plt.plot(x, y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x, y)

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 0])

plt.suptitle("Subplots")

plt.subplot(2, 2, 1)
plt.plot(x, y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 2)
plt.plot(x, y)

plt.subplot(2, 2, 3)
plt.plot(x, y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 4)
plt.plot(x, y)

plt.show()
```

<h4>Legend function:</h4>
<ul>
    <li>It is used to describe the functionality of a line if you plot the multiple lines in a graph.</li>
</ul>

```python
x = np.arange(0, 5, 1)
y1 = (x ** 2)
y2 = x ** 3

plt.plot(x, y1, label = "Squares")
plt.plot(x, y2, label = "Cubes")

plt.legend(title = "Maths", loc = "upper left", framealpha = 1, facecolor = "cyan", edgecolor = "yellow", fancybox = True, shadow = True)

plt.show()
```
