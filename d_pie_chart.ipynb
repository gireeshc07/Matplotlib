<h3>5. Pie charts:</h3>
<ul>
    <li>It draws one piece for each value in the array.</li>
    <li>By default first wedge starts from x-axis and moves counter clockwise.</li>
    <li>Size is determined by comparing all other values using 'x / sum(x)' formula.</li>
</li>
</ul>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
student_values = [15, 25, 12, 8]

plt.figure(figsize = (9, 7))

student_performance = ["Excellent", "Good", "Average", "Poor"]
xx1 = [0.2, 0.0, 0.0, 0.0]
xx2 = ["k", "g", "y", "maroon"]

plt.pie(student_values, startangle = 90, explode = xx1, colors = xx2, wedgeprops = {"edgecolor":"cyan", "linewidth":2}, labels = student_performance)

plt.legend(title = "Performances")

plt.show()
```
