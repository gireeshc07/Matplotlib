<h1>Matplotlib Library:</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h3>3. Bar Chart:</h3>
<ul>
    <li>If we have one categorical and one numerical data, then we can use bar charts. There are two functions to create bar charts:</li>
</ul>
<ol>
    <li>bar() - Vertical Bar</li>
    <li>barh() - Horizontal Bar</li>
</ol>


<h4>Vertical Bar Chart:</h4>

```python
x = ["2020", "2021", "2022", "2023"]
y = [256, 305, 354, 276]

# Creates a Vertical Bar Chart
plt.bar(x, y, color = "r", width = 0.5)

plt.title("College Placements")
plt.xlabel("Y E A R S")
plt.ylabel("No.of Placements")

plt.show()
```

<h4>Horizontal Bar Chart:</h4>

```python
x = ["2020", "2021", "2022", "2023"]
y = [256, 305, 354, 276]
colors = ["r", "k", "y", "c"]

# Creates a Horizontal Bar Chart
plt.barh(x, y, color = colors, height = 0.5)

plt.title("College Placements")
plt.xlabel("Y E A R S")
plt.ylabel("No.of Placements")

plt.show()
```

<h4>Stacked bar charts:</h4>
<ul>
    <li>It represents one category above another category.</li>
</ul>
<h5>1. Vertical Stacked Bar Chart:</h5>

```python
# College Placements
years = ["2020", "2021", "2022", "2023"]
placements = [256, 305, 354, 276]
colors = ["r", "k", "y", "c"]

# plt.figure(figsize = (9, 6))

# Placements of Each Branch
cse = [75, 80, 47, 83]
it = [46, 58, 35, 52]
ece = [67, 59, 72, 62]
eee = [53, 47, 36, 52]

# Specifying the location for bars
e_c_e = [cse[i] + it[i] for i in range(len(cse))]
e_e_e = [e_c_e[i] + ece[i] for i in range(len(cse))]

# Plotting the bar
plt.bar(years, cse, color = "r", width = 0.5, label = "CSE")
plt.bar(years, it, bottom = cse, color = "k", width = 0.5, label = "IT")
plt.bar(years, ece, bottom = e_c_e, color = "g", width = 0.5, label = "ECE")
plt.bar(years, eee, bottom = e_e_e, color = "m", width = 0.5, label = "EEE")

plt.legend(title = "BRANCHES")

plt.title("College Placements")
plt.xlabel("Y E A R S")
plt.ylabel("Placements of Branches")

# Specifying the limit for y-axis
plt.ylim(0, 400)

# Displaying the plot
plt.show()
```

<h5>2. Horizontal Stacked Bar Chart:</h5>

```python
years = ["2020", "2021", "2022", "2023"]
placements = [256, 305, 354, 276]
colors = ["r", "k", "y", "c"]

# plt.figure(figsize = (9, 6))

cse = [75, 80, 47, 83]
it = [46, 58, 35, 52]
ece = [67,59,72,62]
eee = [53, 47, 36, 52]

e_c_e = [cse[i] + it[i] for i in range(len(cse))]
e_e_e = [e_c_e[i] + ece[i] for i in range(len(cse))]

plt.barh(years, cse, color = "r", height = 0.5, label = "CSE")
plt.barh(years, it, left = cse, color = "k", height = 0.5, label = "IT")
plt.barh(years, ece, left = e_c_e, color = "g", height = 0.5, label = "ECE")
plt.barh(years, eee, left = e_e_e, color = "m", height = 0.5, label = "EEE")

plt.legend(title = "BRANCHES")

plt.title("College Placements")
plt.xlabel("Y E A R S")
plt.ylabel("Placements of Branches")

plt.xlim(0, 400)

plt.show()
```

<h4>Grouped bar chart or multiple bar chart</h4>
<ul>
    <li>It plots the bars one after another.</li>
</ul>
<h5>1. Vertical Grouped Bar Charts:</h5>

```python
years = ["2020", "2021", "2022", "2023"]
placements = [256, 305, 354, 276]
colors=["r", "k", "y", "c"]

plt.figure(figsize = (9, 6))

cse = [75, 80, 47, 83]
it = [46, 58, 35, 52]
ece = [67, 59, 72, 62]
eee = [53, 47, 36, 52]
w = 0.2

# locations
w1 = np.arange(len(years))
x = [i + w for i in w1]
y = [i + w for i in x]
z = [i + w for i in y]

# grouped bar
plt.bar(w1, cse, width = w, color = "r", label = "CSE")
plt.bar(x, it, width = w, color = "k", label = "IT")
plt.bar(y, ece, width = w, color = "g", label = "ECE")
plt.bar(z, eee, width = w, color = "m", label = "EEE")

plt.legend(title = "BRANCHES")

plt.title("College Placements")
plt.xlabel("Y E A R S")
plt.ylabel("Placements of Branches")

plt.xticks(w1 + w + w / 2, years)
plt.ylim(0, 130)

plt.show()
```

<h5>2. Horizontal Grouped Bar Charts:</h5>

```python
years = ["2020", "2021", "2022", "2023"]
placements = [256, 305, 354, 276]
colors = ["r", "k", "y", "c"]

plt.figure(figsize = (7, 6))

cse = [75, 80, 47, 83]
it = [46, 58, 35, 52]
ece = [67, 59, 72, 62]
eee = [53, 47, 36, 52]
w = 0.2

# locations
w1 = np.arange(len(years))
x = [i + w for i in w1]
y = [i + w for i in x]
z = [i + w for i in y]

# grouped bar
plt.barh(w1, cse, height = w, color = "r", label = "CSE")
plt.barh(x, it, height = w, color = "k", label = "IT")
plt.barh(y, ece, height = w, color = "g", label = "ECE")
plt.barh(z, eee, height = w, color = "m", label = "EEE")

plt.legend(title = "BRANCHES")

plt.title("College Placements")
plt.xlabel("Y E A R S")
plt.ylabel("Placements of Branches")

plt.yticks(w1 + w + w / 2, years)
plt.xlim(0, 130)

plt.show()
```

<h3>4. Histograms:</h3>
<ul>
    <li>It is a graph showing the frequency distributions.</li>
    <li>It is a graph showing the no.of observations within each given intervals.</li>
</ul>

```python
obs = [90, 50, 40, 60, 55, 44, 30, 32, 34, 84, 69, 62, 88, 99]
intv = [0, 39, 59, 79, 89, 100]
plt.hist(obs, intv, histtype = "stepfilled", rwidth = 0.7, facecolor = "k")

plt.title("Histograms")
plt.xlabel("Intervals")
plt.ylabel("Observations")

plt.xticks([0, 39, 59, 79, 89, 100])
plt.show()
```
